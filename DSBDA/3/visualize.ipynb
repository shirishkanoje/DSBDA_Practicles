```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('global_air_quality_data_10000.csv')
```

```python
df.head()
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
# plot
numerical_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
sns.heatmap(numerical_df.corr(), cmap='YlGnBu', annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# line chart
sns.lineplot(data=df, x='Country', y='Temperature' )
plt.show()
```

```python
# bar plot
plt.figure(figsize=(15, 6))
sns.barplot(data=df, x='Country', y='Temperature')
plt.show()
```

```python
# scatter plot
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='Humidity', y='Temperature',hue='Temperature')
plt.show()
```

```python
# pie chart
plt.figure(figsize=(10,6))
plt.pie(df['Country'].value_counts(), labels=df['Country'].value_counts().index, autopct='%1.1f%%')
plt.title('Country Distribution')
plt.show()
```

```python
# histogram
plt.figure(figsize=(10,6))
sns.histplot(data=df, x= 'Temperature', bins=10, kde=True)
plt.title('Temperature Distribution')
plt.show()
```

```python
# boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Country', y='Temperature')
plt.title('Temperature distribution by Country')
plt.show()
```

```python
# violin plot
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Country', y='Temperature')
plt.title('Temperature distribution by Country')
plt.show()
```
